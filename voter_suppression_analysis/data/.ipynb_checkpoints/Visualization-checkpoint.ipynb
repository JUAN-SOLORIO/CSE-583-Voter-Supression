# Visualization Demostration for Voter-Suppression

For the project we will be calling both modules **processing.py** and **visualization.py**.

```python
# importing both the Processing and Visualization modules
import processing
import visualization as viz
```

```python
# creating data frame for sex_race demographics
sexracedf = processing.combine_sexrace_data()
sexrace_dataframe = processing.homogenize_sexrace_data(sexracedf)
```

```python
sexrace_dataframe.head()
```

```python
# creating age demographics data frame
agedf = processing.combine_age_data()
age_dataframe = processing.homogenize_age_data(agedf)
```

```python
age_dataframe.head()
```

```python
# creating map vizual for age
map_voted = viz.us_map_chart(age_dataframe, map_value='Percent Voted',
             map_title='Percent Voted')

map_registered = viz.us_map_chart(age_dataframe, map_value='Percent Registered',
             map_title='Percent Registered')
```

**Altair** allows for us to use '**|**' and '**&**' for horizontally joining and vertically joining individual altair charts. 
In this case we will use the '**|**' to combine the *Voted* and *Registered* maps.

```python
map_voted | map_registered
```

We can then save a chart by using the "**.save(name.filetype)** function from Altair, in our case we are interested in saving *HTML* files for ease of use.

```python
# saving a map as html
map_voted.save('../figures/Voting_Map.html')
```

```python
# turnout scatter over bars counts/histogram visual for age 
turnout_age = viz.scatter_bar_turnout(df_in=age_dataframe,x_value='Percent Registered:Q',y_value='Percent Voted:Q',
               color_variable='restrictive_id_laws:N',title='Age Groups',
               y_title='Percent Voted', x_title='Percent Registered',
                         select_dropdown='age')
```

```python
turnout_age.save('../figures/Voting_Chart.html')
turnout_age
```

```python
map2_registered = viz.us_map_chart(sexrace_dataframe, map_value='Percent Registered',
             map_title='Percent Registered')

map2_voted = viz.us_map_chart(sexrace_dataframe, map_value='Percent Voted',
             map_title='Percent Voted')
```

```python
map2_voted | map2_registered
```

```python
turnout_sexrace = viz.scatter_bar_turnout(df_in=sexrace_dataframe,x_value='Percent Registered:Q',y_value='Percent Voted:Q',
               color_variable='restrictive_id_laws:N',title='Demographics',
               y_title='Percent Voted', x_title='Percent Registered',
                         select_dropdown='sexrace')
```

```python
turnout_sexrace.save('../figures/Voting_SexRace_Chart.html')
turnout_sexrace
```

```python
(turnout_age | turnout_sexrace)|(map2_voted & map2_registered)
```

```python
# saving Dashboard
((turnout_age | turnout_sexrace) | (map2_voted & map2_registered)).save('../figures/Voter_Turnout_Dashboard.html')
```
